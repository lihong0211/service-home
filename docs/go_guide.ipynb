## 1. 基础语法与变量

```go
package main

import "fmt"

func main() {
	// var 显式声明
	var count int = 42
	var ratio float64 = 2.718
	var lang string = "Golang"
	var enabled bool = true

	// 短变量声明（仅函数内可用）
	score := 88
	e := 2.71828

	// 零值
	var zeroI int
	var zeroS string
	fmt.Printf("零值: int=%d, string=%q\n", zeroI, zeroS)

	// 多重赋值与交换
	u, v := 100, 200
	u, v = v, u
	fmt.Printf("count=%d ratio=%.2f lang=%s enabled=%v\n", count, ratio, lang, enabled)
	fmt.Printf("score=%d e=%.5f 交换后 u=%d v=%d\n", score, e, u, v)
}
```

```go
package main

import "fmt"

func main() {
	// 整数类型
	var i8 int8 = 127
	var u32 uint32 = 1<<32 - 1
	num := 2_000_000
	bin := 0b1111_0000
	oct := 0o755
	hex := 0xDEAD

	// 浮点与复数
	f32 := float32(1.5)
	f64 := 1.41421356
	z := complex(1, 2)
	fmt.Println(real(z), imag(z), z+3i)

	// byte 与 rune
	var b byte = 'B'
	var r rune = '文'
	fmt.Printf("%T %d | %T %d | num=%d | bin=%d oct=%d hex=%d\n", i8, i8, u32, u32, num, bin, oct, hex)
	fmt.Printf("byte=%c(%d) rune=%c(%d)\n", b, b, r, r)
	_, _ = f32, f64
}
```

```go
package main

import "fmt"

// 常量与 iota（包级声明）
const Tau = 6.28318
const (
	CodeOK = 200
	CodeErr = 500
)

type Level int

const (
	LevelLow Level = iota
	LevelMid
	LevelHigh
)

type Flag uint

const (
	FlagA Flag = 1 << iota
	FlagB
	FlagC
)

func main() {
	fmt.Println(Tau, CodeOK, CodeErr)
	fmt.Println(LevelLow, LevelMid, LevelHigh)
	f := FlagA | FlagC
	fmt.Printf("Flag: %b  A=%v B=%v C=%v\n", f, f&FlagA != 0, f&FlagB != 0, f&FlagC != 0)
}
```

```go
package main

import "fmt"

func main() {
	// 函数是一等公民
	square := func(x int) int { return x * x }
	fmt.Println(square(5))
	apply := func(f func(int) int, x int) int { return f(x) }
	fmt.Println(apply(square, 6))
	result := func(a, b int) int { return a + b }(3, 4)
	fmt.Println(result)

	

	// 循环变量捕获（修复：用局部变量）
	funcs := make([]func(), 3)
	for i := 0; i < 3; i++ {
		j := i
		funcs[i] = func() { fmt.Print(j, " ") }
	}
	for _, f := range funcs { f() }
	fmt.Println()
}
```

```go
package main

import "fmt"

func cleanup() {
	defer fmt.Println("defer 1")
	defer fmt.Println("defer 2")
	defer fmt.Println("defer 3")
	fmt.Println("函数体")
}

func deferValue() {
	x := 0
	defer fmt.Println("值参数 x =", x)
	defer func() { fmt.Println("闭包 x =", x) }()
	x = 10
}

func doubleResult(x int) (result int) {
	defer func() { result *= 2 }()
	result = x
	return
}

func main() {
	cleanup()
	fmt.Println("---")
	deferValue()
	fmt.Println("---")
	fmt.Println("doubleResult(5) =", doubleResult(5))
}
```

## 8. 方法与接口

```go
package main

import "fmt"

type Box struct{ W, H int }

func (b Box) Area() int { return b.W * b.H }

func (b *Box) Scale(f int) {
	b.W *= f
	b.H *= f
}

type Shaper interface {
	Area() int
}

func main() {
	box := Box{W: 3, H: 4}
	fmt.Println(box.Area())
	box.Scale(2)
	fmt.Println(box)
	var s Shaper = &box
	fmt.Println(s.Area())
}
```

```go
package main

import (
	"fmt"
	"math"
)

type Circle struct{ Radius float64 }
type Rectangle struct{ Width, Height float64 }

func (c Circle) Area() float64      { return math.Pi * c.Radius * c.Radius }
func (c Circle) Perimeter() float64 { return 2 * math.Pi * c.Radius }
func (c *Circle) Scale(f float64)   { c.Radius *= f }
func (r Rectangle) Area() float64  { return r.Width * r.Height }
func (r Rectangle) Perimeter() float64 { return 2 * (r.Width + r.Height) }

type Shape interface {
	Area() float64
	Perimeter() float64
}

func printShape(s Shape) {
	fmt.Printf("%T: 面积=%.2f 周长=%.2f\n", s, s.Area(), s.Perimeter())
}

func main() {
	c := Circle{Radius: 5}
	r := Rectangle{Width: 4, Height: 3}
	printShape(c)
	printShape(r)
	c.Scale(2)
	fmt.Println("缩放后 Radius =", c.Radius)
	shapes := []Shape{Circle{3}, Rectangle{2, 5}}
	total := 0.0
	for _, s := range shapes { total += s.Area() }
	fmt.Printf("总面积: %.2f\n", total)
}
```

## 9. 错误处理

```go
package main

import (
	"errors"
	"fmt"
)

func divide(a, b float64) (float64, error) {
	if b == 0 {
		return 0, errors.New("除数不能为 0")
	}
	return a / b, nil
}

func main() {
	res, err := divide(10, 3)
	if err != nil {
		fmt.Println("错误:", err)
		return
	}
	fmt.Printf("10/3 = %.2f\n", res)
	res2, err2 := divide(10, 0)
	if err2 != nil {
		fmt.Println("错误:", err2)
	} else {
		fmt.Println(res2)
	}
}
```

```go
package main

import (
	"errors"
	"fmt"
)

var ErrNegative = errors.New("不能为负数")

func sqrtPos(n int) (int, error) {
	if n < 0 {
		return 0, ErrNegative
	}
	return n * n, nil
}

func main() {
	r, err := sqrtPos(5)
	if err != nil {
		fmt.Println(err)
		return
	}
	fmt.Println("5^2 =", r)
	r2, err2 := sqrtPos(-1)
	if err2 != nil {
		fmt.Println("错误:", err2)
	}
	_ = r2
}
```

```go
package main

import (
	"errors"
	"fmt"
)

var ErrNotFound = errors.New("记录不存在")
var ErrPermission = errors.New("权限不足")

type AppError struct { Code int; Message string; Err error }
func (e *AppError) Error() string {
	if e.Err != nil { return fmt.Sprintf("[%d] %s: %v", e.Code, e.Message, e.Err) }
	return fmt.Sprintf("[%d] %s", e.Code, e.Message)
}
func (e *AppError) Unwrap() error { return e.Err }

func findUser(id int) error {
	if id <= 0 { return &AppError{400, "无效ID", ErrNotFound} }
	if id == 999 { return fmt.Errorf("findUser: %w", ErrPermission) }
	return nil
}

func main() {
	err := findUser(-1)
	if errors.Is(err, ErrNotFound) { fmt.Println("未找到:", err) }
	var appErr *AppError
	if errors.As(err, &appErr) { fmt.Printf("AppError Code: %d\n", appErr.Code) }
	fmt.Println(errors.Unwrap(err))
	err2 := findUser(999)
	if errors.Is(err2, ErrPermission) { fmt.Println("权限错误:", err2) }
}
```

```go
package main

import "fmt"

func safeDiv(a, b int) (result int, err error) {
	defer func() {
		if r := recover(); r != nil { err = fmt.Errorf("panic: %v", r) }
	}()
	result = a / b
	return
}

func main() {
	res, err := safeDiv(10, 0)
	fmt.Println(res, err)
	res2, err2 := safeDiv(10, 2)
	fmt.Println(res2, err2)
}
```

```go
package main

import (
	"fmt"
	"sync"
	"time"
)

func main() {
	ch := make(chan int)
	go func() { ch <- 42 }()
	fmt.Println(<-ch)
	buffered := make(chan string, 3)
	buffered <- "a"; buffered <- "b"; buffered <- "c"
	fmt.Println(<-buffered, <-buffered, <-buffered)
	numCh := make(chan int, 5)
	for i := 1; i <= 5; i++ { numCh <- i }
	close(numCh)
	for n := range numCh { fmt.Print(n, " ") }
	fmt.Println()
	producer := func(out chan<- int, n int) {
		for i := 0; i < n; i++ { out <- i }
		close(out)
	}
	pipeCh := make(chan int, 5)
	var wg sync.WaitGroup
	wg.Add(1)
	go func() {
		defer wg.Done()
		for v := range pipeCh { fmt.Print(v, " ") }
	}()
	producer(pipeCh, 5)
	wg.Wait()
	fmt.Println()
}
```

```go
package main

import "fmt"

func generate(nums ...int) <-chan int {
	out := make(chan int, len(nums))
	for _, n := range nums { out <- n }
	close(out)
	return out
}

func squareCh(in <-chan int) <-chan int {
	out := make(chan int)
	go func() {
		for n := range in { out <- n * n }
		close(out)
	}()
	return out
}

func main() {
	for v := range squareCh(squareCh(generate(2, 3, 4))) { fmt.Print(v, " ") }
	fmt.Println()
}
```

## 11. sync 并发原语

```go
package main

import (
	"fmt"
	"sync"
	"sync/atomic"
	"time"
)

func main() {
	var wg sync.WaitGroup
	for i := 0; i < 3; i++ {
		wg.Add(1)
		go func(id int) {
			defer wg.Done()
			time.Sleep(time.Duration(id) * 20 * time.Millisecond)
			fmt.Printf("goroutine %d 完成\n", id)
		}(i)
	}
	wg.Wait()
	fmt.Println("全部完成")
}
```

```go
package main

import (
	"fmt"
	"sync"
	"sync/atomic"
)

type SafeCounter struct { mu sync.Mutex; count int }
func (c *SafeCounter) Inc() { c.mu.Lock(); defer c.mu.Unlock(); c.count++ }
func (c *SafeCounter) Value() int { c.mu.Lock(); defer c.mu.Unlock(); return c.count }

func main() {
	counter := &SafeCounter{}
	var wg sync.WaitGroup
	for i := 0; i < 500; i++ {
		wg.Add(1)
		go func() { defer wg.Done(); counter.Inc() }()
	}
	wg.Wait()
	fmt.Println("计数器:", counter.Value())
	var ac int64
	for i := 0; i < 500; i++ { wg.Add(1); go func() { defer wg.Done(); atomic.AddInt64(&ac, 1) }() }
	wg.Wait()
	fmt.Println("atomic:", atomic.LoadInt64(&ac))
	var sm sync.Map
	sm.Store("name", "Alice")
	val, _ := sm.Load("name")
	fmt.Println(val)
}
```

```go
package main

import (
	"fmt"
	"sync"
)

type Singleton struct{ Value int }
var ( instance *Singleton; once sync.Once )

func GetInstance() *Singleton {
	once.Do(func() { instance = &Singleton{Value: 42}; fmt.Println("初始化单例") })
	return instance
}

func main() {
	var wg sync.WaitGroup
	for i := 0; i < 5; i++ {
		wg.Add(1)
		go func() { defer wg.Done(); GetInstance() }()
	}
	wg.Wait()
	fmt.Println(GetInstance().Value)
}
```

## 12. Context

```go
package main

import (
	"context"
	"fmt"
	"time"
)

type ctxKey string
const requestIDKey ctxKey = "requestID"

func doWork(ctx context.Context) error {
	select {
	case <-time.After(100 * time.Millisecond): return nil
	case <-ctx.Done(): return ctx.Err()
	}
}

func main() {
	ctx, cancel := context.WithCancel(context.Background())
	go func() { time.Sleep(30 * time.Millisecond); cancel() }()
	fmt.Println("WithCancel:", doWork(ctx))
	ctx2, cancel2 := context.WithTimeout(context.Background(), 30*time.Millisecond)
	defer cancel2()
	fmt.Println("WithTimeout:", doWork(ctx2))
	ctx3 := context.WithValue(context.Background(), requestIDKey, "req-123")
	if id, ok := ctx3.Value(requestIDKey).(string); ok { fmt.Printf("请求 ID: %s\n", id) }
}
```

## 13. 泛型（Go 1.18+）

```go
package main

import "fmt"

func First[T any](s []T) T {
	if len(s) == 0 {
		var zero T
		return zero
	}
	return s[0]
}

func main() {
	fmt.Println(First([]int{10, 20, 30}))
	fmt.Println(First([]string{"a", "b", "c"}))
}
```

```go
package main

import "fmt"

type Number interface {
	~int | ~int64 | ~float64
}

func Sum[T Number](nums []T) T {
	var total T
	for _, n := range nums { total += n }
	return total
}

func Map[T, U any](s []T, f func(T) U) []U {
	result := make([]U, len(s))
	for i, v := range s { result[i] = f(v) }
	return result
}

func Filter[T any](s []T, f func(T) bool) []T {
	var result []T
	for _, v := range s { if f(v) { result = append(result, v) } }
	return result
}

func main() {
	fmt.Println(Sum([]int{1, 2, 3, 4, 5}))
	fmt.Println(Map([]int{1, 2, 3}, func(x int) int { return x * 2 }))
	fmt.Println(Filter([]int{1, 2, 3, 4, 5}, func(x int) bool { return x%2 == 0 }))
}
```

```go
package main

import "fmt"

type Stack[T any] struct{ items []T }
func (s *Stack[T]) Push(v T) { s.items = append(s.items, v) }
func (s *Stack[T]) Pop() (T, bool) {
	if len(s.items) == 0 { var zero T; return zero, false }
	item := s.items[len(s.items)-1]
	s.items = s.items[:len(s.items)-1]
	return item, true
}
func (s *Stack[T]) Len() int { return len(s.items) }

type Ordered interface{ ~int | ~float64 | ~string }
func Min[T Ordered](a, b T) T { if a < b { return a }; return b }

func main() {
	st := &Stack[int]{}
	st.Push(1); st.Push(2); st.Push(3)
	for st.Len() > 0 { v, _ := st.Pop(); fmt.Print(v, " ") }
	fmt.Println()
	fmt.Println(Min(3, 5), Min("apple", "banana"))
}
```

## 15. 反射

```go
package main

import (
	"fmt"
	"reflect"
)

type Tag struct {
	ID   int    `json:"id" validate:"required"`
	Name string `json:"name" validate:"min=2"`
}

func inspectStruct(v interface{}) {
	t := reflect.TypeOf(v)
	val := reflect.ValueOf(v)
	if t.Kind() == reflect.Ptr { t = t.Elem(); val = val.Elem() }
	fmt.Printf("类型: %s\n", t.Name())
	for i := 0; i < t.NumField(); i++ {
		f := t.Field(i)
		fmt.Printf("  %s %s json:%s validate:%s\n", f.Name, f.Type, f.Tag.Get("json"), f.Tag.Get("validate"))
	}
}

func main() {
	x := 42
	v := reflect.ValueOf(x)
	fmt.Println(v, reflect.TypeOf(x), reflect.TypeOf(x).Kind())
	y := 100
	pv := reflect.ValueOf(&y).Elem()
	if pv.CanSet() { pv.SetInt(200) }
	fmt.Println(y)
	inspectStruct(Tag{1, "Alice"})
	a, b := []int{1, 2, 3}, []int{1, 2, 3}
	fmt.Println(reflect.DeepEqual(a, b))
}
```

## 16. 高级模式

```go
package main

import (
	"fmt"
	"time"
)

type Server struct { host string; port int; timeout time.Duration; tls bool }
type Option func(*Server)
func WithHost(host string) Option { return func(s *Server) { s.host = host } }
func WithPort(port int) Option { return func(s *Server) { s.port = port } }
func WithTLS(b bool) Option { return func(s *Server) { s.tls = b } }

func NewServer(opts ...Option) *Server {
	s := &Server{host: "localhost", port: 8080, timeout: 30 * time.Second}
	for _, opt := range opts { opt(s) }
	return s
}

func main() {
	s1 := NewServer()
	fmt.Printf("%+v\n", s1)
	s2 := NewServer(WithHost("0.0.0.0"), WithPort(443), WithTLS(true))
	fmt.Printf("%+v\n", s2)
}
```

```go
package main

import (
	"fmt"
	"strings"
)

type QueryBuilder struct { table string; fields []string; where string; limit int }
func NewQuery(table string) *QueryBuilder { return &QueryBuilder{table: table, limit: -1} }
func (q *QueryBuilder) Select(fields ...string) *QueryBuilder { q.fields = fields; return q }
func (q *QueryBuilder) Where(cond string) *QueryBuilder { q.where = cond; return q }
func (q *QueryBuilder) Limit(n int) *QueryBuilder { q.limit = n; return q }
func (q *QueryBuilder) Build() string {
	fields := "*"
	if len(q.fields) > 0 { fields = strings.Join(q.fields, ", ") }
	sql := fmt.Sprintf("SELECT %s FROM %s", fields, q.table)
	if q.where != "" { sql += " WHERE " + q.where }
	if q.limit > 0 { sql += fmt.Sprintf(" LIMIT %d", q.limit) }
	return sql
}

func main() {
	sql := NewQuery("users").Select("id", "name").Where("age > 18").Limit(10).Build()
	fmt.Println(sql)
}
```

## 17. 测试（表驱动）

```go
package main

import "fmt"

func Add(a, b int) int { return a + b }
func Factorial(n int) int {
	if n <= 1 { return 1 }
	return n * Factorial(n-1)
}

func main() {
	fmt.Println("Add(2,3) =", Add(2, 3))
	cases := []struct{ input, want int }{{0, 1}, {1, 1}, {5, 120}, {10, 3628800}}
	for _, c := range cases {
		got := Factorial(c.input)
		if got != c.want { fmt.Printf("FAIL Factorial(%d)=%d want %d\n", c.input, got, c.want) } else { fmt.Printf("PASS Factorial(%d)=%d\n", c.input, got) }
	}
}
```

## 18. 调试与性能

```go
package main

import (
	"fmt"
	"log"
	"runtime"
	"strings"
	"sync"
	"time"
	"bytes"
)

func main() {
	log.SetFlags(log.Ldate | log.Ltime | log.Lshortfile)
	log.Println("标准日志")
	fmt.Printf("Go 版本: %s OS/Arch: %s/%s CPU: %d Goroutine: %d\n", runtime.Version(), runtime.GOOS, runtime.GOARCH, runtime.NumCPU(), runtime.NumGoroutine())
	var m runtime.MemStats
	runtime.ReadMemStats(&m)
	fmt.Printf("堆分配: %d KB\n", m.HeapAlloc/1024)
	withPrealloc := func(n int) []int {
		result := make([]int, 0, n)
		for i := 0; i < n; i++ { result = append(result, i) }
		return result
	}
	start := time.Now()
	r := withPrealloc(10000)
	fmt.Printf("预分配 10000 元素: %v 最后值: %d\n", time.Since(start), r[9999])
	var bufPool = sync.Pool{New: func() interface{} { return new(bytes.Buffer) }}
	buf := bufPool.Get().(*bytes.Buffer)
	buf.Reset(); buf.WriteString("hello pool")
	fmt.Println(buf.String())
	bufPool.Put(buf)
	var sb strings.Builder
	for i := 0; i < 5; i++ { fmt.Fprintf(&sb, "%d,", i) }
	fmt.Println("Builder:", sb.String())
}
```
