```python
score = 85
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'D'
print(grade)

# 三元表达式
label = "pass" if score >= 60 else "fail"
print(label)

# match-case（Python 3.10+）
day = "Tue"
match day:
    case "Sat" | "Sun":
        print("周末")
    case "Mon":
        print("周一!")
    case _:
        print("其他!")
```

```python
# ---------- for 循环 ----------
# 基础遍历
for i in range(5):
    print(i, end=' ')
print()

# enumerate —— 同时获得索引和值
fruits = ['apple', 'banana', 'cherry']
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")

# zip —— 并行遍历多个可迭代对象
names = ['Alice', 'Bob', 'Carol']
scores = [90, 85, 92]
for name, score in zip(names, scores):
    print(f"{name}: {score}")

# for...else（循环正常结束时执行 else）
for n in [2, 3, 5, 7]:
    if n % 2 == 0 and n != 2:
        break
else:
    print("全部为质数")


# ---------- while 循环 ----------
n = 1
while n < 32:
    n *= 2
print(n)   # 32

# break / continue
for i in range(10):
    if i == 7:
        break
    if i % 2 == 0:
        continue
    print(i, end=' ')   # 1 3 5
print()
```

```python
import functools, time

def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f"{func.__name__} 耗时 {elapsed:.6f}s")
        return result
    return wrapper
```

```python
def make_counter():
    count = 0
    def counter():
        nonlocal count
        count += 1
        return count
    return counter
counter = make_counter()
print(counter(), counter(), counter())  # 1 2 3
```

```python
class Animal:
    kingdom = "狮王"
    def __init__(self, name, sound):
        self.name = name
        self._sound = sound
    def speak(self):
        return f"{self.name} says {self._sound}"
class Dog(Animal):
    def __init__(self, name, sound, breed):
        super().__init__(name, "汪汪！")
        self.breed = breed
    def speak(self):
        return f"{self.name}是{self.breed}，{super().speak()}"
        
dog = Dog("旺财", "汪汪！", "拉布拉多")
print(dog.speak())
```

```python
# Python：异常机制
def divide(a, b):
    if b == 0:
        raise ZeroDivisionError("除数不能为 0")
    return a / b

try:
    result = divide(10, 0)
except ZeroDivisionError as e:
    print("错误:", e)
except TypeError as e:
    print("类型错误:", e)
else:
    print("成功:", result)  # 无异常时执行
finally:
    print("always")         # 总是执行
```

```python
import asyncio

async def fetch(id, delay):
    await asyncio.sleep(delay)
    return f"数据{id}完成"

async def main():
    results = await asyncio.gather(
        fetch(1, 0.3),
        fetch(2, 0.2),
        fetch(3, 0.1)
    )
    for r in results: print(r)

await main()
```

```python
import threading

counter = 0
lock = threading.Lock()

def increment(n):
    global counter
    for _ in range(n):
        with lock:
            if counter < 20: print(counter)
            counter += 1

threads = [threading.Thread(target=increment, args = (10000, )) for _ in range(5)]
for t in threads: t.start()
for t in threads: t.join()
print(f"最终计数: {counter}")
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

def fetch_url(url):
    time.sleep(0.1)
    return f"fetched: {url}"

urls = [f"http://example.com/{i}" for i in range(10)]

with ThreadPoolExecutor(max_workers=4) as executor:
    # results = list(executor.map(fetch_url, urls))
    results = [x for x in executor.map(fetch_url, urls)]
print(results[:3])

with ThreadPoolExecutor(max_workers=3) as ex:
    futures = {ex.submit(fetch_url, u): u for u in urls[:5]}
    for future in as_completed(futures):
        print(future.result())

```
